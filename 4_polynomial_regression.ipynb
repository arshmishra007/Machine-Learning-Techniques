```python
# its a method to build the non linear relationship between the dependent and the independent variable
```

```python
# it is used when the data shows a curvey trend
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,10,500)
c = np.random.randn(500)
y = x**3+c*100+100
```

```python
plt.scatter(x,y)
```

```python
# importing the linear regresion library
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
x = x.reshape(-1,1)
```

```python
model.fit(x,y)
```

```python
pred = model.predict(x)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,pred)
```

```python
plt.scatter(x,y)
plt.plot(x,pred,linewidth=4,color='black')
plt.show()
```

```python
# that is why we are getting the bad performance
```

```python
# implementing the polynomial regression
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_model = PolynomialFeatures()
```

```python
poly_x = poly_model.fit_transform(x)
```

```python
poly_x.shape
```

```python
# now fitting the data in LinearRegression
model.fit(poly_x,y)
```

```python
pred = model.predict(poly_x)
```

```python
r2_score(y,pred)
```

```python
#accuracy score increased from 74 to 88 percent
```

```python
plt.scatter(x,y)
plt.plot(x,pred,color='black')
```
