# MULTICOLLINEARITY IN LINEAR REGRESSION

#it arises when we have a large number of features and they are highly correlated with each other that is there p value is greater than 0.05

```python
# let's see the example 1
```

```python
import pandas as pd
```

```python
import statsmodels.api as sm
```

```python
dataset1=pd.read_csv("Advertising.csv")
```

```python
dataset1.columns
```

```python
dataset1.drop("Unnamed: 0",inplace=True,axis=1)
```

```python
dataset1.head()
```

```python
#now seperating the dependent and independent features 
X = dataset1.iloc[:,:-1] #(excluding the last column)
Y = dataset1.iloc[:,-1]  #(including only the last column)
```

```python
X
```

```python
Y
```

```python
#so now our equation for the best fit line will be of the form -> y_hat = m0 + m1x1 +m2x2 +m3x3
# for m0 we will add a constant coefficient

#technique used -> ordinary least square(ols)
X = sm.add_constant(X)
```

```python
X.head()
```

```python
model = sm.OLS(Y,X).fit()
```

```python
model.summary()
```

```python
# 1)check coefficients -> appropriate values
# 2)R squared value  -> close to 1
# 3)stderror value -> acceptable
# 4)p value should be less than 0.05 (newspaper isn't following this value so removing it)
X.drop('newspaper',axis=1,inplace=True)
```

```python
X
```

```python
model = sm.OLS(Y,X).fit()
```

```python
model.summary()
```

```python
#Now we have removed the Multicollinearity
```

# Example 2 for multicollinearity

```python
dataset2 = pd.read_csv("Salary_Data.csv")
```

```python
dataset2.head()
```

```python
#our formula becomes y_hat = mo +m1x1 +m2x2
```

```python
#seperating the dataset into the dependent and the independent features 
X = dataset2.iloc[:,:-1]#independent features 
Y = dataset2.iloc[:,-1]#dependent feature
```

```python
X.head()
```

```python
#now we will add the constant m0
X=sm.add_constant(X)
```

```python
X
```

```python
model2 = sm.OLS(Y,X).fit()
```

```python
model2.summary()
```
