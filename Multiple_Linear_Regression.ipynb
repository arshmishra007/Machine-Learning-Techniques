# Multiple Linear Regression

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset.head()
```

```python
# now seperating the dependent and the independent features
```

```python
X = dataset.iloc[:,:-1]#independent features (excluding the last column)
Y = dataset.iloc[:,-1] #dependent features (including only the last column)
```

```python
#our target will be to find out the profit
```

```python
#state is a categorical variable 
states = pd.get_dummies(dataset['State'],drop_first=True)
```

```python
states
```

```python
X.drop('State',axis=1,inplace=True)
```

```python
X.head()
```

```python
X=pd.concat([X,states],axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
prediction = regression.predict(X_test)
```

```python
prediction
```

```python
#now for the comparison
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,prediction)
```

```python
score
```
